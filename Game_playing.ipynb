```python
# --------------------------------------------------
# MINIMAX ALGORITHM WITH OPTIMAL PATH
# --------------------------------------------------

def minimax(node, is_max):

    # Leaf node
    if isinstance(node, int):
        return node, []

    # MAX player
    if is_max:
        best_value = float('-inf')
        best_path = []

        for name, child in node.items():
            value, path = minimax(child, False)

            if value > best_value:
                best_value = value
                best_path = [name] + path

        return best_value, best_path

    # MIN player
    else:
        best_value = float('inf')
        best_path = []

        for name, child in node.items():
            value, path = minimax(child, True)

            if value < best_value:
                best_value = value
                best_path = [name] + path

        return best_value, best_path


# --------------------------------------------------
# GAME TREE FROM THE IMAGE
# --------------------------------------------------

game_tree = {

    'B': {

        'D': {

            'H': {
                'P': 3,
                'Q': 4
            },

            'I': {
                'R': 2,
                'S': 1
            }
        },

        'E': {

            'J': {
                'T': 7,
                'U': 8
            },

            'K': {
                'V': 9,
                'W': 10
            }
        }
    },

    'C': {

        'F': {

            'L': {
                'X': 2,
                'Y': 11
            },

            'M': {
                'Z': 1,
                'a': 12
            }
        },

        'G': {

            'N': {
                'b': 14,
                'c': 9
            },

            'O': {
                'd': 13,
                'e': 16
            }
        }
    }
}


# --------------------------------------------------
# RUN MINIMAX
# --------------------------------------------------

optimal_value, optimal_path = minimax(game_tree, True)

print("Optimal Value =", optimal_value)
print("Optimal Path = A -> " + " -> ".join(optimal_path))
```

```python
# Alpha-Beta Pruning for given game tree

import math
# Import math module to use infinity values (math.inf, -math.inf)
# These are used to initialize best/worst possible values for MAX and MIN nodes

# Game tree representation (only structure, not values)
# Game tree representation
tree = {

    'A': ['B', 'C'],

    'B': ['D', 'E'],
    'C': ['F', 'G'],

    'D': ['H', 'I'],
    'E': ['J', 'K'],
    'F': ['L', 'M'],
    'G': ['N', 'O'],

    'H': ['P', 'Q'],
    'I': ['R', 'S'],
    'J': ['T', 'U'],
    'K': ['V', 'W'],
    'L': ['X', 'Y'],
    'M': ['Z', 'a'],
    'N': ['b', 'c'],
    'O': ['d', 'e']
}

# Leaf node values
values = {

    'P': 3,
    'Q': 4,

    'R': 2,
    'S': 1,

    'T': 7,
    'U': 8,

    'V': 9,
    'W': 10,

    'X': 2,
    'Y': 11,

    'Z': 1,
    'a': 12,

    'b': 14,
    'c': 9,

    'd': 13,
    'e': 16
}

# Alpha-Beta Pruning function definition
def alpha_beta(node, depth, alpha, beta, maximizingPlayer, path):
    # node → current node in tree
    # depth → level of tree (used for indentation/visualization)
    # alpha → best value that MAX can guarantee so far
    # beta → best value that MIN can guarantee so far
    # maximizingPlayer → True if MAX's turn, False if MIN's turn
    # path → keeps track of traversal path from root to current node

    indent = "  " * depth
    # Creates indentation based on depth for better tree-like printing

    # ---------------- BASE CASE ----------------
    # If node is not in tree dictionary, it is a leaf node
    if node not in tree:
        print(f"{indent}Leaf {node} = {values[node]}")
        # Print leaf node value

        return values[node], path + [node]
        # Return leaf value and updated path including this node

    # ---------------- MAXIMIZING PLAYER ----------------
    if maximizingPlayer:
        value = -math.inf
        # Initialize MAX value to lowest possible number

        best_path = []
        # Store best path leading to optimal value

        print(f"{indent}MAX node {node}")
        # Indicate we are at a MAX node

        # Explore all children of current node
        for child in tree[node]:

            # Recursive call for child node as MIN player
            val, child_path = alpha_beta(
                child, depth + 1, alpha, beta, False, path + [node]
            )

            # If returned value is better than current best, update
            if val > value:
                value = val
                best_path = child_path

            # Update alpha (best value MAX can guarantee)
            alpha = max(alpha, value)

            print(f"{indent}Updated MAX {node}: value={value}, alpha={alpha}, beta={beta}")

            # Pruning condition: if alpha >= beta, stop exploring remaining children
            if alpha >= beta:
                print(f"{indent} Pruning at {node} (alpha >= beta)")
                break

        # Return best value found and corresponding path
        return value, best_path

    # ---------------- MINIMIZING PLAYER ----------------
    else:
        value = math.inf
        # Initialize MIN value to highest possible number

        best_path = []
        # Store best path for MIN player

        print(f"{indent}MIN node {node}")
        # Indicate we are at a MIN node

        # Explore all children of current node
        for child in tree[node]:

            # Recursive call for child node as MAX player
            val, child_path = alpha_beta(
                child, depth + 1, alpha, beta, True, path + [node]
            )

            # If returned value is smaller, update (MIN wants lowest value)
            if val < value:
                value = val
                best_path = child_path

            # Update beta (best value MIN can guarantee)
            beta = min(beta, value)

            print(f"{indent}Updated MIN {node}: value={value}, alpha={alpha}, beta={beta}")

            # Pruning condition: if alpha >= beta, stop exploring
            if alpha >= beta:
                print(f"{indent} Pruning at {node} (alpha >= beta)")
                break

        # Return best value found and corresponding path
        return value, best_path


# ---------------- RUN THE ALGORITHM ----------------

print("\n===== ALPHA-BETA PRUNING TRACE =====\n")
# Print heading for clarity of output

# Start algorithm from root node 'A'
# Initial alpha = -infinity (MAX worst case)
# Initial beta = +infinity (MIN worst case)
# Start with MAX player
result, path = alpha_beta('A', 0, -math.inf, math.inf, True, [])

# ---------------- FINAL OUTPUT ----------------
print("\n===== FINAL RESULT =====")
print("Optimal value at root (A):", result)
# Best achievable value from root node A

print("Optimal path:", " -> ".join(path))
# Path taken to reach optimal decisionn
```
