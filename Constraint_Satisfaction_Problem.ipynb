```python
'''QSN: You are given 2 jugs , a 4 litre one and a 3 litre one , a pump which has unlimited water which you can
 to fill the jug and tnhe ground on which water may be poured. Neither jug has any measuring
 makings on it. How can you get exactly 2 litre of water in the 4 litre jug?'''
```

```python
import collections

def water_jug_problem(jug1_capacity, jug2_capacity, target_amount):
    # State: (water in jug1, water in jug2)
    initial_state = (0, 0)

    # Queue for BFS, storing (state, path_to_state)
    queue = collections.deque([(initial_state, [])])

    # Set to keep track of visited states
    visited = {initial_state}

    while queue:
        current_state, path = queue.popleft()
        jug1_water, jug2_water = current_state

        # Check if target is reached in the specified jug
        if jug1_water == target_amount:
            return path + [f"Target reached: {jug1_water}L in {jug1_capacity}L jug, {jug2_water}L in {jug2_capacity}L jug"]

        # Possible actions:
        next_states = []

        # 1. Fill jug1
        if jug1_water < jug1_capacity:
            next_states.append(((jug1_capacity, jug2_water), f"Fill {jug1_capacity}L jug"))

        # 2. Fill jug2
        if jug2_water < jug2_capacity:
            next_states.append(((jug1_water, jug2_capacity), f"Fill {jug2_capacity}L jug"))

        # 3. Empty jug1
        if jug1_water > 0:
            next_states.append(((0, jug2_water), f"Empty {jug1_capacity}L jug"))

        # 4. Empty jug2
        if jug2_water > 0:
            next_states.append(((jug1_water, 0), f"Empty {jug2_capacity}L jug"))

        # 5. Pour from jug1 to jug2
        if jug1_water > 0 and jug2_water < jug2_capacity:
            pour_amount = min(jug1_water, jug2_capacity - jug2_water)
            next_states.append(((jug1_water - pour_amount, jug2_water + pour_amount), f"Pour from {jug1_capacity}L jug to {jug2_capacity}L jug"))

        # 6. Pour from jug2 to jug1
        if jug2_water > 0 and jug1_water < jug1_capacity:
            pour_amount = min(jug2_water, jug1_capacity - jug1_water)
            next_states.append(((jug1_water + pour_amount, jug2_water - pour_amount), f"Pour from {jug2_capacity}L jug to {jug1_capacity}L jug"))

        for next_state, action_description in next_states:
            if next_state not in visited:
                visited.add(next_state)
                new_path = path + [f"State: {current_state} -> {next_state}. Action: {action_description}"]
                queue.append((next_state, new_path))

    return "No solution found."

# Solve the problem
jug1_capacity = 4
jug2_capacity = 3
target_amount = 2

solution_path = water_jug_problem(jug1_capacity, jug2_capacity, target_amount)

print(f"Problem: Get {target_amount}L in the {jug1_capacity}L jug using a {jug1_capacity}L and a {jug2_capacity}L jug.")
if isinstance(solution_path, list):
    for step in solution_path:
        print(step)
else:
    print(solution_path)
```
